```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features = ['holiday','workingday','weather','temp','humidity','windspeed','hour','weekday','month','year']
x=train[features]
y=train['count']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

LR_model = LinearRegression()
LR_model.fit(x_train, y_train)
pred_LR = LR_model.predict(x_test)
```

```python
def rmsle(y_true, y_pred):
    y_pred = np.maximum(0, y_pred)   
    return np.sqrt(np.mean((np.log1p(y_pred) - np.log1p(y_true))**2))
```

```python
print("Linear Regression RMSLE:", rmsle(y_test, pred_LR))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)

X_train_poly = poly.fit_transform(x_train)
X_val_poly = poly.transform(x_test)

lr_poly = LinearRegression()
lr_poly.fit(X_train_poly, y_train)
pred_lr_poly = lr_poly.predict(X_val_poly)

print("Poly Linear Regression RMSLE:", rmsle(y_test, pred_lr_poly))
```

```python
scaler = StandardScaler(with_mean=False)
X_train_poly_scaled = scaler.fit_transform(X_train_poly)
X_val_poly_scaled = scaler.transform(X_val_poly)

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_poly_scaled, y_train)
pred_ridge = ridge.predict(X_val_poly_scaled)

print("Ridge RMSLE:", rmsle(y_test, pred_ridge))
```

```python
lasso = Lasso(alpha=0.001, max_iter=10000)
lasso.fit(X_train_poly_scaled, y_train)
pred_lasso = lasso.predict(X_val_poly_scaled)
pred_lasso = np.maximum(0, pred_lasso)


print("Lasso RMSLE:", rmsle(y_test, pred_lasso))
```

```python
results = pd.DataFrame({
    'Model': ['Linear', 'Poly Linear', 'Ridge', 'Lasso'],
    'RMSLE': [
        rmsle(y_test, pred_LR),
        rmsle(y_test, pred_lr_poly),
        rmsle(y_test, pred_ridge),
        rmsle(y_test, pred_lasso)
    ]
})
results
```
